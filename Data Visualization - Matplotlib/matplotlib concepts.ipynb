```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sample')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
# Object oriented method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('FIRST PLOT')
axes[1].plot(y,x)
axes[1].set_title('SECOND PLOT')
plt.tight_layout()
plt.show()
```

## Figure size and dpi

```python
fig = plt.figure(figsize=(4,3))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))

axes[0].plot(x,y,'b')
axes[0].set_title('FIRST PLOT')
axes[1].plot(y,x,'r')
axes[1].set_title('SECOND PLOT')
plt.tight_layout()
plt.show()
```

```python
fig.savefig('My_test_plot.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=0)
plt.show()
```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#F5843B',lw=3,alpha=1,ls='--')
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#F5843B',lw=3,alpha=1,ls=':')
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#F5843B',lw=3,alpha=1,ls='steps')
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=1,alpha=1,ls='-',marker='o',markersize=10)
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,alpha=1,ls='-',marker='o',markersize=10
        ,markerfacecolor='red',markeredgewidth=2,markeredgecolor='black')
plt.show()
```

```python
#for 0 to 1 values

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
plt.show()
```
